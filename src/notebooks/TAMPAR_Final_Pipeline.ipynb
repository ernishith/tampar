```python
# calculate difference in results
import pandas as pd

BASELINE_CSV = "/content/tampar/tampering_results_test.csv"
NEW_CSV      = "/content/tampar/tampering_results_test (3).csv"

METRICS = ["accuracy", "precision_binary", "recall_binary", "f1_binary", "roc_auc"]
KEYS = ["predictor", "compare_types"]

def load_results(path: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    # keep only what we need for comparison
    cols = KEYS + METRICS
    return df[cols].copy()

base = load_results(BASELINE_CSV)
new  = load_results(NEW_CSV)

# align rows by keys, then compute deltas
merged = base.merge(new, on=KEYS, how="inner", suffixes=("_base", "_new"))

out = merged[KEYS].copy()
out["accuracy Δ"]   = merged["accuracy_new"] - merged["accuracy_base"]
out["precision Δ"]  = merged["precision_binary_new"] - merged["precision_binary_base"]
out["recall Δ"]     = merged["recall_binary_new"] - merged["recall_binary_base"]
out["f1 Δ"]         = merged["f1_binary_new"] - merged["f1_binary_base"]
out["roc_auc Δ"]    = merged["roc_auc_new"] - merged["roc_auc_base"]

# optional: only show simple_threshold predictor (your files use this) [file:1][file:4]
out = out[out["predictor"] == "simple_threshold"].drop(columns=["predictor"])

# pretty print like the response above
pd.set_option("display.float_format", lambda x: f"{x:+.4f}")
print(out.to_markdown(index=False))

# optional: save for reuse
out.to_csv("comparison_deltas.csv", index=False)

```

### Sample Image comparison before and after attack


```python
import matplotlib.pyplot as plt
import cv2

img_paths = [
    '/content/drive/MyDrive/TAMPAR_DATA/tampar/test/base/id_00_20230512_102251.jpg',
    '/content/drive/MyDrive/TAMPAR_DATA/tampar/adversarial_attacks/test/base_adv/id_00_20230512_102251_fgsm.jpg',
    '/content/drive/MyDrive/TAMPAR_DATA/tampar/test/base/id_00_20230512_102302.jpg',
    '/content/drive/MyDrive/TAMPAR_DATA/tampar/adversarial_attacks/test/base_adv/id_00_20230512_102302_pgd.jpg'
]

image_titles = [
    'Clean Image 1',
    'FGSM Image 1',
    'Clean Image 2',
    'PGD Image 2'
]

# --- 2. Read and Prepare Images ---
images = []
for path in img_paths:
    img = cv2.imread(path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    images.append(img_rgb)


fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Flatten axes array for easy iteration
axes = axes.flatten()

for i, img in enumerate(images):
    axes[i].imshow(img)
    axes[i].set_title(image_titles[i])
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
